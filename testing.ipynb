```python
import matplotlib.pyplot as plt
import numpy as np
import librosa as lbr
import pandas as pd
```

```python
amp, sampla_rate = lbr.load("signal_2.wav")
print(amp)
print(len(amp))
```

```python
pd.Series(amp).plot(figsize=(10, 5),title="Normal Signal",xlabel="time",ylabel="Amplitude")
plt.show()
```

```python
ft = np.fft.fft(amp)
freq = np.fft.fftfreq(len(amp))
plt.plot(freq, ft.real**2 + ft.imag**2)
plt.title("Signal Spectrum")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```

## Letter A 

```python
a_letter=amp[18000:32500]
print(a_letter)
len(a_letter)
plt.subplot(2,1,1)
pd.Series(a_letter).plot(figsize=(10, 5),title="Letter A",xlabel="time",ylabel="Amplitude")

plt.subplot(2,1,2)
ft = np.fft.fft(a_letter)
freq = np.fft.fftfreq(len(a_letter))
plt.plot(freq, ft.real**2+ft.imag**2)
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.show()
```
